```python
boxscores_advanced = pd.read_pickle('boxscores_advanced_team_2000.pkl')
```

```python
boxscores_basic = pd.read_pickle('raw_games_5yrs.pkl')
```

```python
features_adv_team = ['TEAM_ID', 'OFF_RATING', 'DEF_RATING', 'NET_RATING', 
            'AST_PCT', 'AST_TOV', 'OREB_PCT', 'DREB_PCT', 'REB_PCT', 
            'TM_TOV_PCT', 'EFG_PCT', 'TS_PCT', 'PACE', 'POSS', 'HOME_TEAM']
non_features_columns = ['GAME_ID', 'TEAM_ABBREVIATION', 'PLUS_MINUS']
df_features = ['GAME_ID', 'TEAM_ABBREVIATION', 'OFF_RATING', 'DEF_RATING', 'NET_RATING', 
            'AST_PCT', 'AST_TOV', 'OREB_PCT', 'DREB_PCT', 'REB_PCT', 
            'TM_TOV_PCT', 'EFG_PCT', 'TS_PCT', 'PACE', 'POSS', 'HOME_TEAM', 'PLUS_MINUS']
```

```python
boxscores_advanced['GAME_ID'] = boxscores_advanced['GAME_ID'].astype('int32')
```

```python
temp_df = boxscores_basic[['GAME_ID', 'TEAM_ABBREVIATION', 'PLUS_MINUS', 'HOME_TEAM']]
temp_df
```

```python
boxscores_advanced = boxscores_advanced.merge(temp_df, on=['GAME_ID', 'TEAM_ABBREVIATION'])
```

```python
boxscores_advanced.reset_index(drop=True, inplace=True)
```

```python
boxscores_advanced
```

```python
boxscores_advanced_sorted = boxscores_advanced.sort_values(['GAME_ID', 'HOME_TEAM'])
```

```python
boxscores_advanced_sorted = boxscores_advanced_sorted[df_features]
boxscores_advanced_sorted.info()
```

```python
bsa_away = boxscores_advanced_sorted.iloc[::2].reset_index(drop=True)
columns = bsa_away.columns.tolist()
```

```python
bsa_home = boxscores_advanced_sorted.iloc[1::2].reset_index(drop=True)
```

```python
columns_away = {}
columns_home = {}
for column in columns:
    if column == 'GAME_ID':
        continue
    columns_away[column] = column + '_a'
    columns_home[column] = column + '_h'
```

```python
bsa_away.rename(columns=columns_away, inplace=True)
bsa_home.rename(columns=columns_home, inplace=True)
```

```python
merged_df = bsa_home.merge(bsa_away, on=['GAME_ID'])
```

```python
merged_df.info()
```

```python
merged_df.drop('PLUS_MINUS_a', axis=1, inplace=True)
```

```python
merged_df.rename(columns={'PLUS_MINUS_h': 'PLUS_MINUS_h-a'}, inplace=True)
```

```python
merged_df.columns
```

```python
X_features_num = ['OFF_RATING_h', 'DEF_RATING_h',
       'NET_RATING_h', 'AST_PCT_h', 'AST_TOV_h', 'OREB_PCT_h', 'DREB_PCT_h',
       'REB_PCT_h', 'TM_TOV_PCT_h', 'EFG_PCT_h', 'TS_PCT_h', 'PACE_h',
       'POSS_h', 'HOME_TEAM_h',
       'OFF_RATING_a', 'DEF_RATING_a', 'NET_RATING_a', 'AST_PCT_a',
       'AST_TOV_a', 'OREB_PCT_a', 'DREB_PCT_a', 'REB_PCT_a', 'TM_TOV_PCT_a',
       'EFG_PCT_a', 'TS_PCT_a', 'PACE_a', 'POSS_a', 'HOME_TEAM_a']

X_features_cat = ['TEAM_ABBREVIATION_h', 'TEAM_ABBREVIATION_a']

X_features_all = ['TEAM_ABBREVIATION_h', 'OFF_RATING_h', 'DEF_RATING_h',
       'NET_RATING_h', 'AST_PCT_h', 'AST_TOV_h', 'OREB_PCT_h', 'DREB_PCT_h',
       'REB_PCT_h', 'TM_TOV_PCT_h', 'EFG_PCT_h', 'TS_PCT_h', 'PACE_h',
       'POSS_h', 'HOME_TEAM_h', 'TEAM_ABBREVIATION_a',
       'OFF_RATING_a', 'DEF_RATING_a', 'NET_RATING_a', 'AST_PCT_a',
       'AST_TOV_a', 'OREB_PCT_a', 'DREB_PCT_a', 'REB_PCT_a', 'TM_TOV_PCT_a',
       'EFG_PCT_a', 'TS_PCT_a', 'PACE_a', 'POSS_a', 'HOME_TEAM_a']
```

```python
X = merged_df[X_features_all]
y = merged_df['PLUS_MINUS_h-a']
```

```python
scaler = MinMaxScaler()
X[X_features_num] = scaler.fit_transform(X[X_features_num])
```

```python
ohe = OneHotEncoder(sparse=False)
ohe.fit(X[X_features_cat])
cols = [str(team) +'_h' for team in ohe.categories_[0]] + [str(team) +'_a' for team in ohe.categories_[1]]
```

```python
X[cols]=ohe.transform(X[X_features_cat])
```
